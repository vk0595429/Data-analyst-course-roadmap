```python
import pandas as pd
```

```python
salary=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
salary.head()
```

```python
salary.info()
```

```python
salary.describe()
```

```python
salary.columns
```

```python
# Define tARGet y and features x
y=salary['Salary']
x=salary[['Experience Years']]
```

```python
#Apply train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
#select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# train model ( fit modeln)
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#line y= c(intercept) + m(slope)expericence + error
#y= 27020 +9343experinence + error
#linear dsta model or mechanical data model
```

```python
#prediction
y_pred=model.predict(x_test)
```

```python
x_test
```

```python
#accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
